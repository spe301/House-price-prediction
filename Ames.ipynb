```python
import pandas as pd
df = pd.read_csv(r'C:\Users\aacjp\OneDrive\Desktop\train.csv')
d = pd.read_csv(r'C:\Users\aacjp\OneDrive\Desktop\test.csv')
df
```

```python
quantifiable_df = df.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition'], axis='columns')
quantifiable_d = d.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition'], axis='columns')
```

```python
continuous_df = quantifiable_df.drop(['LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'PavedDrive'], axis='columns')
continuous_d = quantifiable_d.drop(['LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'PavedDrive'], axis='columns')
```

```python
use_later = continuous_d['Id']
```

```python
df1 = continuous_df.drop(['Id', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual',
'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical',                      
'KitchenQual', 'Functional',  'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'], axis='columns')
test = continuous_d.drop(['Id', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual',
'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical',                      
'KitchenQual', 'Functional',  'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'], axis='columns')
```

```python
floors = []
for i in range(1460):
    if df1['HouseStyle'][i][0] != 'S':
        floors.append(int(df1['HouseStyle'][i][0]))
    else:
        floors.append(2)
    
df1['HouseStyle'] = floors
```

```python
stories = []
for i in range(1459):
    if test['HouseStyle'][i][0] != 'S':
        stories.append(int(test['HouseStyle'][i][0]))
    else:
        stories.append(2)
    
test['HouseStyle'] = stories
```

```python
lf_mean = round(df1['LotFrontage'].mean() ,1)
mva_mean = round(df1['MasVnrArea'].mean() ,1)
gyb_mean = round(df1['GarageYrBlt'].mean())
lfa = round(test['LotFrontage'].mean() ,1)
mvaa = round(test['MasVnrArea'].mean() ,1)
bfsa = round(test['BsmtFinSF1'].mean() ,1)
bfsa2 = round(test['BsmtFinSF2'].mean() ,1)
busa = round(test['BsmtUnfSF'].mean() ,1)
tbsa = round(test['TotalBsmtSF'].mean() ,1)
bfba = round(test['BsmtFullBath'].mean() ,1)
bhba = round(test['BsmtHalfBath'].mean() ,1)
gyba = round(test['GarageYrBlt'].mean())
gca = round(test['GarageCars'].mean() ,1)
gaa = round(test['GarageArea'].mean() ,1)
```

```python
df1['LotFrontage'] = df1['LotFrontage'].fillna(lf_mean)
df1['MasVnrArea'] = df1['MasVnrArea'].fillna(mva_mean)
df1['GarageYrBlt'] = df1['GarageYrBlt'].fillna(gyb_mean)
test['LotFrontage'] = test['LotFrontage'].fillna(lfa)
test['MasVnrArea'] = test['MasVnrArea'].fillna(mvaa)
test['BsmtFinSF1'] = test['BsmtFinSF1'].fillna(bfsa)
test['BsmtFinSF2'] = test['BsmtFinSF2'].fillna(bfsa2)
test['BsmtUnfSF'] = test['BsmtUnfSF'].fillna(busa)
test['TotalBsmtSF'] = test['TotalBsmtSF'].fillna(tbsa)
test['BsmtFullBath'] = test['BsmtFullBath'].fillna(bfba)
test['BsmtHalfBath'] = test['BsmtHalfBath'].fillna(bhba)
test['GarageYrBlt'] = test['GarageYrBlt'].fillna(gyba)
test['GarageCars'] = test['GarageCars'].fillna(gca)
test['GarageArea'] = test['GarageArea'].fillna(gaa)
```

```python
y = df1.SalePrice
x = df1.drop(['SalePrice'], axis='columns')
x
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def get_prices(x, y, coefs):
    model = LinearRegression().fit(x, y)
    prices = []
    for i in range(1459):
        prices.append(model.predict([list(coefs.iloc[i])])[0])
    return prices
```

```python
get_prices(x, y, test)
```

```python
price_predictions = get_prices(x, y, test)
id_list = list(use_later)
output_dict = {}
for i in range(1459):
    output_dict[id_list[i]] = price_predictions[i]
```

```python
id_df = pd.DataFrame(id_list)
price_df = pd.DataFrame(price_predictions)
final_df = pd.concat([id_df, price_df], axis='columns')
final_df.to_csv('C:/Users/aacjp/Documents/Ames.csv')
```
